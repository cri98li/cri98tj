```python
nMov_RIG7 = pd.read_csv("Vehicle search multiple rappresentation.csv")
nMov_RIG6 = pd.read_csv("Vehicle search multiple rappresentation precision 6.csv")
```

```python
nMov_RIG6.describe().apply(lambda s: s.apply('{0:.5f}'.format))
```

```python
nMov_RIG7.describe().apply(lambda s: s.apply('{0:.5f}'.format))
```

```python
df_scatter_veh = nMov_RIG6[(nMov_RIG6.sliding == 3) & (nMov_RIG6.k==10) & (nMov_RIG6.nclust!=10)]

df_scatter_veh
```

```python
plt.rcParams['figure.figsize'] = [7.0, 5.5]
plt.rcParams['figure.dpi'] = 120
plt.rcParams.update({'font.size': 14})

plt.scatter(df_scatter_anim.nclust, df_scatter_anim.acc, label="Animals dataset")
plt.scatter(df_scatter_veh.nclust, df_scatter_veh.acc, label="Vehicles dataset")

plt.legend()

plt.ylabel("Accuracy", fontsize=18)
plt.xlabel("k", fontsize=18)
plt.tick_params(axis='both', which='major', labelsize=18)

plt.ylim(ymin=0, ymax=1)
plt.xlim(xmin=0)

plt.title("Accuracy score", fontsize=22)


plt.savefig("TrAC results.pdf", bbox_inches='tight')
```

```python
plt.rcParams['figure.figsize'] = [7.0, 5.5]
plt.rcParams['figure.dpi'] = 120
plt.rcParams.update({'font.size': 14})

plt.scatter(df_scatter_anim.nclust, df_scatter_anim.t/1000/60, label="Animals dataset")
plt.scatter(df_scatter_veh.nclust, df_scatter_veh.t/1000/60, label="Vehicles dataset")

plt.legend(loc=5)

plt.ylabel("Time (min)", fontsize=18)
plt.xlabel("k", fontsize=18)
plt.tick_params(axis='both', which='major', labelsize=18)

plt.ylim(ymin=0)
plt.xlim(xmin=0)

plt.title("Execution Time", fontsize=22)


plt.savefig("TrAC results time.pdf", bbox_inches='tight')
```
