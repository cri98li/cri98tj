```python
import time
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from tqdm.notebook import tqdm
import math
from datetime import timedelta, datetime
import plotly.express as px
from geolib import geohash
```

```python
df_animals_euclidea = pd.read_csv("../experiments/Test euclidea animals.csv")
df_animals_IRD = pd.read_csv("../experiments/Test IRD animals.csv")
df_animals_DTW = pd.read_csv("../experiments/Test DTW animals.csv")
```

```python
plt.plot(df_animals_euclidea.precision, df_animals_euclidea.acc, 'o', label="ED")
plt.plot(df_animals_IRD.precision, df_animals_IRD.acc, 'o', label="IRD")
plt.plot(df_animals_DTW.precision, df_animals_DTW.acc, 'o', label="DTW")

plt.legend()

plt.legend()
plt.ylim(ymin=0, ymax=1)
#plt.xlim(xmin=0)
fig = plt.gcf()
fig.set_size_inches(5.5, 4)
fig.set_dpi(300)

plt.xlabel("Geohash precision", fontsize=16)
plt.ylabel("Accuracy", fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=16)
plt.savefig("animals distance.pdf", bbox_inches='tight')

plt.show()
```

```python
plt.plot(df_animals_euclidea.precision, df_animals_euclidea.t/1000, 'o', label="ED")
plt.plot(df_animals_IRD.precision, df_animals_IRD.t/1000, 'o', label="IRD")
plt.plot(df_animals_DTW.precision, df_animals_DTW.t/1000, 'o', label="DTW")

ax = plt.gca()
#ax.set_yscale('log')

plt.legend()

plt.legend()
plt.yscale('log')
fig = plt.gcf()
plt.ylim(ymin=8, ymax=10**4)
fig.set_size_inches(5.5, 4)
fig.set_dpi(300)
plt.tick_params(axis='both', which='major', labelsize=16)

plt.xlabel("Geohash precision", fontsize=16)
plt.ylabel("Time (s)", fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=16)
plt.savefig("animals distance time.pdf", bbox_inches='tight')

plt.show()
```

```python
df_vehicles_euclidea = pd.read_csv("../experiments/Test euclidea vehicles.csv")
df_vehicles_IRD = pd.read_csv("../experiments/Test IRD vehicles.csv")
df_vehicles_DTW = pd.read_csv("../experiments/Test DTW vehicles.csv")
```

```python
plt.plot(df_vehicles_euclidea.precision[2:], df_vehicles_euclidea.acc[2:], 'o', label="ED")
plt.plot(df_vehicles_IRD.precision[2:], df_vehicles_IRD.acc[2:], 'o', label="IRD")
plt.plot(df_vehicles_DTW.precision[:-1], df_vehicles_DTW.acc[:-1], 'o', label="DTW")

plt.legend()

plt.legend()
plt.ylim(ymin=0, ymax=1)
#plt.xlim(xmin=0)
fig = plt.gcf()
fig.set_size_inches(5.5, 4)
fig.set_dpi(300)
plt.xlabel("Geohash precision", fontsize=16)
plt.ylabel("Accuracy", fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=16)
plt.xticks([5, 6, 7])
plt.savefig("vehicles distance.pdf", bbox_inches='tight')

plt.show()
```

```python
plt.plot(df_vehicles_euclidea.precision[2:], df_vehicles_euclidea.t[2:]/1000, 'o', label="ED")
plt.plot(df_vehicles_IRD.precision[2:], df_vehicles_IRD.t[2:]/1000, 'o', label="IRD")
plt.plot(df_vehicles_DTW.precision[:-1], df_vehicles_DTW.t[:-1]/1000, 'o', label="DTW")

ax = plt.gca()
ax.set_yscale('log')

plt.legend()

plt.legend()
plt.ylim(ymin=10, ymax=10**4)
#plt.xlim(xmin=0)
fig = plt.gcf()
fig.set_size_inches(5.5, 4)
fig.set_dpi(300)
plt.xlabel("Geohash precision", fontsize=16)
plt.ylabel("Time (s)", fontsize=16)
plt.xticks([5, 6, 7])
plt.tick_params(axis='both', which='major', labelsize=16)
plt.savefig("vehicles distance time.pdf", bbox_inches='tight')

plt.show()
```
